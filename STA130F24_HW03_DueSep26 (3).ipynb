````python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np

# Load the penguins dataset from an online source
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(url)

# Function to add annotations
def add_annotations(fig, species, data):
    mean = data.mean()
    median = data.median()
    std = data.std()
    iqr = data.quantile(0.75) - data.quantile(0.25)
    min_val = data.min()
    max_val = data.max()
    
    # Add mean and median lines
    fig.add_hline(y=mean, line=dict(color='blue', dash='dash'), annotation_text='Mean', row=species, col=1)
    fig.add_hline(y=median, line=dict(color='green', dash='dash'), annotation_text='Median', row=species, col=1)
    
    # Add range and IQR rectangles
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor='lightgrey', opacity=0.3, line_width=0, row=species, col=1)
    fig.add_vrect(x0=mean-2*std, x1=mean+2*std, fillcolor='lightblue', opacity=0.3, line_width=0, row=species, col=1)
    fig.add_vrect(x0=median-iqr/2, x1=median+iqr/2, fillcolor='lightgreen', opacity=0.3, line_width=0, row=species, col=1)

# Create subplots for each species
fig = make_subplots(rows=3, cols=1, shared_xaxes=True, subplot_titles=df['species'].unique())

# Plot histograms and add annotations
for i, species in enumerate(df['species'].unique(), 1):
    species_data = df[df['species'] == species]['flipper_length_mm']
    fig.add_trace(go.Histogram(x=species_data, name=species), row=i, col=1)
    add_annotations(fig, i, species_data)

# Update layout
fig.update_layout(height=900, width=700, title_text="Flipper Length Distribution by Species with Annotations")

fig.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the penguins dataset from an online source
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(url)

# Set the style
sns.set_style("whitegrid")

# Create a figure with three subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Function to add annotations
def add_annotations(ax, data):
    mean = data.mean()
    median = data.median()
    std = data.std()
    iqr = data.quantile(0.75) - data.quantile(0.25)
    min_val = data.min()
    max_val = data.max()
    
    # Add mean and median lines
    ax.axvline(mean, color='blue', linestyle='--', label='Mean')
    ax.axvline(median, color='green', linestyle='--', label='Median')
    
    # Add range and IQR rectangles
    ax.axvspan(min_val, max_val, color='lightgrey', alpha=0.3, label='Range')
    ax.axvspan(mean-2*std, mean+2*std, color='lightblue', alpha=0.3, label='±2 Std Dev')
    ax.axvspan(median-iqr/2, median+iqr/2, color='lightgreen', alpha=0.3, label='IQR')

# Plot KDE and add annotations for each species
for ax, species in zip(axes, df['species'].unique()):
    species_data = df[df['species'] == species]['flipper_length_mm'].dropna()
    sns.kdeplot(species_data, ax=ax, fill=True)
    add_annotations(ax, species_data)
    ax.set_title(species)
    ax.legend()

# Update layout
fig.suptitle("Flipper Length Distribution by Species with Annotations", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
````

Box Plots(box-and-whisker plots)- display the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. They also highlight outliers.
Pros:
Summarizes data: Provides a clear summary of the data distribution.
Identifies outliers: Easily shows outliers.
Compares distributions: Effective for comparing distributions across different groups.
Cons:
Limited detail: Does not show the distribution shape in detail.
No frequency information: Does not provide information about the frequency of data points.

Histograms- bar graphs that represent the frequency distribution of a dataset. The data is divided into bins, and the height of each bar represents the number of data points in each bin.
Pros:
Detailed distribution: Shows the shape of the data distribution.
Frequency information: Provides information about the frequency of data points.
Cons:
Bin size sensitivity: The appearance can change significantly with different bin sizes.
Less effective for small datasets: Can be misleading if the dataset is too small.

Kernel Density Estimators (KDEs)-smooth curves that estimate the probability density function of a dataset. They use a kernel function to smooth the data points and create a continuous distribution.
Pros:
Smooth distribution: Provides a smooth and continuous estimate of the data distribution.
Adjustable bandwidth: The bandwidth parameter can be adjusted to control the smoothness.
Cons:
Bandwidth sensitivity: The choice of bandwidth can significantly affect the appearance.
Computationally intensive: More computationally intensive than histograms and box plots.

````python
import pandas as pd
import plotly.express as px

# Load the dataset
bn = pd.read_csv('https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv')

# Make identical boy and girl names distinct
bn['name'] = bn['name'] + " " + bn['sex']

# Rank the names by percent within each year
bn['rank'] = bn.groupby('year')['percent'].rank(ascending=False)

# Sort the DataFrame by name and year
bn = bn.sort_values(['name', 'year'])

# Create the percent change column
bn['percent change'] = bn['percent'].diff()
new_name = [True] + list(bn.name[:-1].values != bn.name[1:].values)
bn.loc[new_name, 'percent change'] = bn.loc[new_name, 'percent']

# Sort the DataFrame by year
bn = bn.sort_values('year')

# Restrict to "common" names
bn = bn[bn.percent > 0.001]

# Create the scatter plot
fig = px.scatter(bn, 
                 x="percent change", 
                 y="rank", 
                 animation_frame="year", 
                 animation_group="name", 
                 size="percent", 
                 color="sex", 
                 hover_name="name", 
                 size_max=50, 
                 range_x=[-0.005, 0.005])

# Reverse the y-axis to put rank 1 on top
fig.update_yaxes(autorange='reversed')

# Display the plot
fig.show(renderer="png")  # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````
